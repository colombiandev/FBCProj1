```python
# Calculate daily standard deviations 
value_std = value_daily_return.std()
value_std
```

```python
# Calculate annualized standard deviation (252 trading days)
value_volatility = value_daily_return.std() * np.sqrt(252)
value_volatility = value_volatility
value_volatility
```

```python
# Annualized Sharpe Ratios

value_sharpe_ratios = (value_daily_return.mean() * 252) / (value_daily_return.std() * np.sqrt(252))
value_sharpe_ratios.sort_values()
```

```python
# Visualize the sharpe ratios as a bar plot
value_sharpe_ratios.hvplot.bar(title="Value ETF Sharpe Ratios")
```
